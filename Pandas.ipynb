```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Hz=1.5
ans=2*np.pi*Hz
x=np.arange(0,10,.1)
y=np.sin(ans*x)
z=np.cos(ans*x)
```

```python
df=pd.DataFrame({'Time':x,'YVals':y,'ZVals':z})
df
```

```python
Extract_data=df[["YVals","ZVals"]][10:16]
Extract_data
```
